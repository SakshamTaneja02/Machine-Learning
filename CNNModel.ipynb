```python
import tensorflow
```

```python
(train_features,train_targets),(test_features,test_targets)=tensorflow.keras.datasets.fashion_mnist.load_data()
```

```python
type(train_features)
```

```python
train_features.shape
#here it means that model has 60000 images of the size 28*28
```

train features contain the images of objects that are used to train the model. 
all these are stored in the form of array

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(train_features[i])
plt.show()
```

```python
from keras.utils.np_utils import to_categorical
```

```python
train_targets=to_categorical(train_targets)
```

```python
train_targets.shape
```

# Step 1: specifying the architecture 

```python
from keras.layers import Dense,Flatten
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

# compile the model

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

# train the model

```python
model.fit(train_features,train_targets,epochs=10)
```

# test the model

```python
predictions=model.predict(test_features)
```
